### Object Detection - Selective search

```python
import matplotlib.pyplot as plt
%matplotlib inline

import numpy as np
import os
import sys
import cv2
import tensorflow as tf
```

```python
# Загрузка графического файла
img = cv2.imread('img_pack\hight-st-image.jpg')
# Задаем размерность изображения
newHeight = 200
newWidth = int(img.shape[1]*newHeight / img.shape[0])
# Применяем новые размеры к изображению
img = cv2.resize(img, (newWidth, newHeight))
```

```python
# Подключаем режим мультипроцессорности
cv2.setUseOptimized(True)
cv2.setNumThreads(8)
```

```python
# Создаем объект Selective searct segmentation
sss = cv2.ximgproc.segmentation.createSelectiveSearchSegmentation()
# Задаем изображение
sss.setBaseImage(img)
```

```python
# Выбираем метод Selective Search
sss.switchToSelectiveSearchQuality()
# Запускаем алгоритм
rects = sss.process() # Данный методв вернет области с объектами
print(rects.shape)
```

```python
# Выводим изображение на экран
plt.figure(figsize = (10, 10))
plt.imshow(img)
plt.show()
```

```python
# Задаем способы обработки изображения
numRects = 20 # Кол-во регионов для поиска
# Копия исходной картинки
imOut = img.copy()

for i, rect in enumerate(rects):
    if (i < numRects):
        x, y, w, h = rect
        cv2.rectangle(imOut,
                      (x, y),
                      (x + w, y + h),
                      (0, 255, 0),
                      1,
                      cv2.LINE_AA)
    else:
        break
```

```python
# Выводим изображение с регионами
plt.figure(figsize = (10, 10))
plt.imshow(imOut)
plt.show()
```
