```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
```

```python
# Загрузка датасета с погодными данными
url = "https://raw.githubusercontent.com/vega/vega/master/docs/data/seattle-weather.csv"
data = pd.read_csv(url)
data.head()
```

```python
# Описательная статистика
data.describe()
```

```python
# Нормализация данных MinMax Scaler
minmax_scaler = MinMaxScaler() # Создаем объект
# Обучение MinMax Scaler
data_norm = minmax_scaler.fit_transform(
    np.array(data[['precipitation', 'temp_max', 'temp_min', 'wind']]))
```

```python
data_norm[1,:]
```

```python
data_norm[:1]
```

```python
# Преобразуем обратно в датафрейм
data_norm_df = pd.DataFrame(
    data_norm, columns = ['precipitation', 'temp_max', 'temp_min', 'wind'])
data_norm_df.head()
```

```python
data_norm_df.describe()
```

```python
# Применение стандартизации StandardScaler
std_scaler = StandardScaler() # Создаем объект
# Обучение StandardScaler
data_std = std_scaler.fit_transform(
    np.array(data[['precipitation', 'temp_max', 'temp_min', 'wind']]))
```

```python
data_std_df = pd.DataFrame(
    data_std, columns = ['precipitation', 'temp_max', 'temp_min', 'wind'])
data_std_df.head()
```

```python
data_std_df.describe()
```
