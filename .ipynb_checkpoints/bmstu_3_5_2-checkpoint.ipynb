```python
# Получить количество уникальных значений по каждому полю
from operator import itemgetter
sorted([(col, ml[col].nunique()) for col in ml.columns], key = itemgetter(1))
```

```python
# Преобразование к категориальным данным
ml_with_cat = ml.copy()
ml_with_cat['recclass'] = ml_with_cat['recclass'].astype('category')
```

```python
ml_with_cat.info()
```

```python
%%timeit
# Определим среднюю массу по каждой категории
ml.groupby('recclass')['mass'].mean().to_frame()
```

```python
%%timeit
ml_with_cat.groupby('recclass')['mass'].mean().to_frame()
```

```python
# Сортировка
ml.sort_values(by=['recclass', 'mass'], ascending=[True, False]).head(5)
```

```python
# Извлечение значений полей
ml[ml['name']=='Abee']['mass'].mean()
```

```python
# Получение различных значений поля
ml.fall.value_counts()
```

```python
# Применение метода map
d = {'Found':1, 'Fell':0}
ml['status'] = ml['fall'].map(d)
ml
```
