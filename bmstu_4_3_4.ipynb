```python
%matplotlib inline
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as sps
import pandas as pd
import numpy as np
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
```

```python
# Загрузка датасета
iris = sns.load_dataset('iris')
iris.head()
```

```python
# Отделение исходных признаков от цели предсказания
X = iris.drop(['species'], axis=1).values
y = iris['species'].values
```

```python
# Разделение наборов данных
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
print(f"Кол-во строк обучающего набора данных {X_train.shape[0]}")
print(f"Кол-во строк тестового набора данных {X_test.shape[0]}")
```

```python
# Разведовательный анализ
g = sns.PairGrid(iris, hue='species', height=4, diag_sharey=False)
g.map_lower(sns.kdeplot)
g.map_upper(plt.scatter)
g.map_diag(sns.histplot, kde=True)
```

```python
# Создание Pipeline (Нормализация данных + Классификатор на основе SVC)
pipe = Pipeline([('scaler', StandardScaler()), ('svc', SVC())])
pipe.fit(X_train, y_train)
```

```python
# Оценка качества модели
score = float('{:.3f}'.format(pipe.score(X_test, y_test)))
print(f"Точность модели: {score * 100}%")
```

```python
# Проверка
pred = pipe.predict(X_test[10].reshape(1, -1))
print(f"Предсказанное значение: {pred[0]}")
print(f"Истинное значение: {y_test[10]}")
```
