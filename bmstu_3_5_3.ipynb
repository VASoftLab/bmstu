```python
import pandas as pd
# Объединение таблиц
left = pd.DataFrame(
    {
        "A": ["A0", "A1", "A2"],
        "B": ["B0", "B1", "B2"]
    },
    index=["K0", "K1", "K2"])
right = pd.DataFrame(
    {
        "C": ["C0", "C1", "C2"],
        "D": ["B0", "B1", "B2"]
    },
    index=["K0", "K2", "K3"])
result = left.join(right)
result
```

```python
result = left.join(right, how="outer")
result
```

```python
# Аналогичны способ объединения таблиц
result = pd.merge(left, right, left_index = True, right_index = True,
                 how = "outer")
result
```

```python
# Объединение по ключу и индексу
left = pd.DataFrame(
    {
        "A": ["A0", "A1", "A2", "A3"],
        "B": ["B0", "B1", "B2", "B3"],
        "key": ["K0", "K1", "K0", "K1"]
    }
)
right = pd.DataFrame(
    {
        "C": ["C0", "C1"],
        "D": ["D0", "D1"]
    },
    index = ["K0", "K1"]
)
result = left.join(right, on="key")
result
```
