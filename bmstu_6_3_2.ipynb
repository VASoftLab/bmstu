### Метод c-mens

```python
import sklearn
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from fcmeans import FCM
```

```python
# Задаем массив
X = np.array([[0, 0],
            [1, 0],
            [0, 1],
            [1, 1],
            [2, 1],
            [1, 2],
            [2, 2],
            [3, 2],
            [6, 6],
            [7, 6],
            [8, 6],
            [6, 7],
            [7, 7],
            [8, 7],
            [9, 7],
            [7, 8],
            [8, 8],
            [9, 8],
            [8, 9],
            [9, 9],])
```

```python
fcm = FCM(n_clusters = 2)
fcm.fit(X)
```

```python
fcm_centers = fcm.centers
fcm_labels = fcm.predict(X)
```

```python
# plot result
f, axes = plt.subplots(1, 2, figsize = (11, 5), dpi = 300)
axes[0].scatter(X[:, 0], X[:, 1])
axes[1].scatter(X[:, 0], X[:, 1], c = fcm_labels)
axes[1].scatter(fcm_centers[:, 0], fcm_centers[:, 1], marker = "+", s = 200, c = 'b')
plt.show()
```

```python
from __future__ import division, print_function
import skfuzzy as fuzz
```

```python
colors = ['b', 'orange', 'g', 'r', 'c', 'm', 'y', 'k', 'Brown', 'ForestGreen']
# Определяем центры кластеров
centers = [[4, 2], [1, 7], [5, 6]]
# Определяем sigmas для X И Y соответственно
sigmas = [[0.8, 0.3], [0.3, 0.5], [1.1, 0.7]]
# Генерация тестовых данных
np.random.seed(111)
xpts = np.zeros(1)
ypts = np.zeros(1)
lbls = np.zeros(1)
for i, ((xmu, ymu), (xsigma, ysigma)) in enumerate(zip(centers, sigmas)):
    xpts = np.hstack((xpts, np.random.standard_normal(200) * xsigma + xmu))
    ypts = np.hstack((ypts, np.random.standard_normal(200) * ysigma + xmu))
    lbls = np.hstack((lbls, np.ones(200) * i))
    
# Визуализация
fig0, ax0 = plt.subplots(dpi = 300)
for lbl in range(3):
    ax0.plot(xpts[lbls == lbl], ypts[lbls == lbl], '.', color = colors[lbl])
ax0.set_title('Тестовые данные: 200 точек 3 кластера')
```

```python
# Позволяет лучший результат по числу кластеров
fig1, ax1 = plt.subplots(3, 3, figsize = (8, 8), dpi = 300)
all_data = np.vstack((xpts, ypts))
fpcs = []

for ncenters, ax in enumerate(ax1.reshape(-1), 2):
    cntr, u, u0, d, jm, p, fpc = fuzz.cluster.cmeans(
    all_data, ncenters, 2, error = 0.005, maxiter = 1000, init = None)
    
    fpcs.append(fpc)
    
    cluster_memnership = np.argmax(u, axis = 0)
    for j in range(ncenters):
        ax.plot(xpts[cluster_memnership == j],
               ypts[cluster_memnership == j], '.', color = colors[j])
        
    for pt in cntr:
        ax.plot(pt[0], pt[1], 'rs')
        
    ax.set_title(f"Centers = {ncenters}; FPC = {fpc:.2}")
    ax.axis("off")

fig1.tight_layout()
```

```python
fig2, ax2 = plt.subplots(dpi = 300)
ax2.plot(np.r_[2:11], fpcs)
ax2.set_xlabel('Кол-во центров')
ax2.set_ylabel('Fuzzy partition coefficient')
```
