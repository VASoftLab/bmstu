### Метод k-средний

```python
import sklearn
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
# Задаем массив
X = np.array([[0, 0],
            [1, 0],
            [0, 1],
            [1, 1],
            [2, 1],
            [1, 2],
            [2, 2],
            [3, 2],
            [6, 6],
            [7, 6],
            [8, 6],
            [6, 7],
            [7, 7],
            [8, 7],
            [9, 7],
            [7, 8],
            [8, 8],
            [9, 8],
            [8, 9],
            [9, 9],])
```

```python
X
```

```python
# Нанесем точки на плоскость
plt.scatter(X[:, 0], X[:, 1])
```

```python
# Предварительно видно, что кол-во кластеров 2
k_means = KMeans(n_clusters = 2)
k_means.fit(X)
```

```python
k_meancntr = k_means.cluster_centers_
print(f"Печать центров кластеров: {k_meancntr}")
print(f"Печать меток кластеров: {k_means.labels_}")
```

```python
# Визуализация центров
plt.scatter(X[:, 0], X[:, 1], c = k_means.labels_, marker = '^', cmap = 'rainbow')
plt.scatter(k_means.cluster_centers_[:, 0], k_means.cluster_centers_[:, 1], color = 'black')
```

#### Ирисы Фишера

```python
from sklearn import datasets
```

```python
iris = datasets.load_iris()
print(iris)
```

```python
X = iris.data
y = iris.target
dfY = pd.DataFrame(y)
print(dfY[0].unique())
```

```python
# Три различных класса
k_means_iris = KMeans(n_clusters = 3).fit(X)
print(k_means_iris.cluster_centers_)
```

```python
# Визуализация
fig = plt.figure(figsize = (10, 4), dpi = 300)
# Метки которые нашел кластеризатор
ax = fig.add_subplot(121)
plt.scatter(X[:, 0], X[:, 1], c = k_means_iris.labels_, marker = 'o')
# Истинные значения
ax = fig.add_subplot(122)
plt.scatter(X[:, 0], X[:, 1], c = y, marker = 'o')
```
