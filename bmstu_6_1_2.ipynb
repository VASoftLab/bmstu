### Обучение на неразмеченных данных

```python
import pandas as pd
import numpy as np

from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs

import matplotlib.pyplot as plt
```

```python
# Установка генератора случайных чисел
np.random.seed(40)
```

```python
# Генерация случайных кластеров
X, y = make_blobs(n_samples = 200, n_features = 2,
                 centers = 4, cluster_std = 0.7,
                 shuffle = True)

plt.title("Сгенерированные данные")
plt.scatter(X[:, 0], X[:, 1], c = y)
plt.show()
```

```python
X # Неразмеченные данные
```

### Реализация алгоритма k-средних

```python
kmeans = KMeans(n_clusters = 4, init = 'random',
               n_init = 10, max_iter = 300, tol = 1e-04)
kmeans_prediction = kmeans.fit_predict(X)
```

```python
plt.title("Результаты кластеризации")
plt.scatter(X[:, 0], X[:, 1], c = kmeans_prediction)
plt.scatter(*kmeans.cluster_centers_.T,
           marker = "*", s = 200, color = 'r')
```

##### Использование метода локтя для получения количества кластеров

```python
distortions = []
for i in range(1, 7):
    kmeans = KMeans(n_clusters = i,
                   init = 'random',
                   n_init = 10, max_iter = 300,
                   tol = 1e-04)
    kmeans_prediction = kmeans.fit_predict(X)
    plt.title(f"Результат кластеризации при {i} кластерах")    
    plt.scatter(X[:, 0], X[:, 1], c = kmeans_prediction)
    plt.scatter(*kmeans.cluster_centers_.T,
                marker = "*", s = 200, color = 'r')
    plt.show()
    distortions.append(kmeans.inertia_)
```

```python
# Применение метода локтя
plt.plot(range(1, 7), distortions, marker = 'o')
plt.xlabel('Количество кластеров')
plt.ylabel('Разброс')
plt.show()
# Очеидно, что разница между 4 и 5 минимальна,
# принимаем что кластеров 4.
# Поиск останавливается, когда резкое падение кончается
```
