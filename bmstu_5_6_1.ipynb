### Доверительные интервалы

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.utils import resample
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
%matplotlib inline
```

```python
# Загрузка данных
data = pd.read_csv('D:\\SourceCode\\jupyter-notebook\\bmstu\\pima-indians-diabetes.csv')
data.head()
```

```python
values = data.values
# Алгоритм Bootstrap
n_iteration = 1000
n_size = int(len(data)*0.5)
# Запуск Bootstrap
stats = [] # Создаем пуустой список
for i in range(n_iteration):
    train = resample(values, n_samples = n_size)
    test = np.array([x for x in values if x.tolist() not in train.tolist()])
    # Обучение модели
    model = DecisionTreeClassifier()
    model.fit(train[:,:-1], train[:,-1])
    # Прогноз
    predict = model.predict(test[:,:-1])
    score = accuracy_score(test[:,-1], predict)
    print(f"{i}\t{score}")
    stats.append(score)
    
stats[:10]
```

```python
# Гистограмма точности
plt.hist(stats)
plt.show()

# Расчет доверительного интервала
alpha = 0.95
p = ((1-alpha)/2)*100
lower = max(0.0, np.percentile(stats, p))

p = (alpha + (1-alpha)/2)*100
upper = min(1, np.percentile(stats, p))

print(alpha*100, lower*100, upper*100)
# Наша модель работает с точностью 69% при этом
# нижняя граница доверительного интервала 64% верхняя 73%
```
