### Support Vector Machines
Метод опорных векторов для бинарной классификации

```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
```

```python
# Генерируем синтетические данные
from sklearn.datasets import make_blobs
X, y = make_blobs(n_samples = 50, centers = 2,
                  random_state = 0, cluster_std = 0.60)
plt.scatter(X[:, 0], X[:, 1], c = y, s = 50, cmap = 'winter')
```

```python
# Строим разделяющую гиперплоскость
xfit = np.linspace(-1, 3.5)
fig = plt.figure(figsize = (5, 5), dpi = 150)
plt.scatter(X[:, 0], X[:, 1], c = y, s = 50, cmap = 'winter')
for m, b, d in [(1, 0.65, 0.33), (0.5, 1.6, 0.55), (-0.2, 2.9, 0.2)]:
    yfit = m * xfit + b
    plt.plot(xfit, yfit, '-k')
    plt.fill_between(xfit, yfit - d, yfit + d, edgecolor = 'none', color = '#AAAAAA', alpha = 0.4)
plt.xlim(-1, 3.5)
plt.grid(color = 'b', linestyle = '--', linewidth = 0.5)
# Разделение возможно различными способами (ручное разделение)
```

```python
from sklearn.svm import SVC
svm_clf = SVC(kernel = 'linear')
svm_clf.fit(X, y)
```

```python
# Функция для построения границы решателей SVM
def plot_svc_decision_function(model, ax = None, plot_support = True):
    if ax is None:
        ax = plt.gca()
    xlim = ax.get_xlim()
    ylim = ax.get_ylim()
    
    x = np.linspace(xlim[0], xlim[1], 30)
    y = np.linspace(ylim[0], ylim[1], 30)
    Y, X = np.meshgrid(y, x)
    xy = np.vstack([X.ravel(), Y.ravel()]).T
    P = model.decision_function(xy).reshape(X.shape)
    ax.contour(X, Y, P, colors = 'k',
              levels = [-1, 0, 1], alpha = 0.5,
              linestyles = ['--', '-', '--'])
    
    # plot support vectors
    if plot_support:
        ax.scatter(model.support_vectors_[:, 0],
                   model.support_vectors_[:, 1],
                  s = 200, linewidth = 0.75,
                  facecolors = 'none', edgecolors = 'b')
    
    ax.set_xlim(xlim)
    ax.set_ylim(ylim)
```

```python
fig = plt.figure(figsize = (5, 3), dpi = 150)
plt.scatter(X[:, 0], X[:, 1], c = y, s = 50, cmap = 'winter')
plot_svc_decision_function(svm_clf)
# То, что обведено кружочками, является опорными векторами
```

```python
# Интерактивное взаимодействие с параметрами метода
from ipywidgets import interact

def plot_svm(N, svc_C):
    X, y = make_blobs(n_samples = 200, centers = 2,
                  random_state = 0, cluster_std = 0.60)
    X = X[: N]
    y = y[: N]
    svm = SVC(kernel='linear', C = svc_C)
    svm.fit(X, y)
    fig = plt.figure(figsize = (5, 3), dpi = 150)
    plt.scatter(X[:, 0], X[:, 1], c = y, s = 50, cmap = 'winter')
    plt.xlim(-1, 4)
    plt.ylim(-1, 6)
    plot_svc_decision_function(svm, plt.gca())
    plt.scatter(svm.support_vectors_[:, 0],
                svm.support_vectors_[:, 1],
                s = 200, linewidth = 0.75,
                facecolors = 'none', edgecolors = 'red')
    
interact(plot_svm, N = [50, 100, 150, 200], svc_C = (0.1, 10))
```

```python
# Исследование влияния ядра SVM
from sklearn.datasets import make_circles
X, y = make_circles(100, factor = 0.1, noise = 0.1)
clf = SVC(kernel = 'linear').fit(X, y)
fig = plt.figure(figsize = (10, 10), dpi = 150)
plt.scatter(X[:, 0], X[:, 1], c = y, s = 50, cmap = 'winter')
plot_svc_decision_function(clf)
# Разделения не произошло
```

```python
# Добавляем новую размерность
r = np.exp(-(X[:, 0]**2 + X[:, 1]**2))       
```

```python
from mpl_toolkits import mplot3d
```

```python
def plot_3D(elev = 30, azim = 30):
    fig = plt.figure(figsize = (10, 10), dpi = 150)
    ax = plt.subplot(projection = '3d')
    ax.scatter3D(X[:, 0], X[:, 1], r, c = y, s = 50, cmap = 'winter')
    ax.view_init(elev = elev, azim = azim)
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_zlabel('r')
    
interact(plot_3D, elev = [-90, -45, 45, 90], azim = (-180, 180))
```

```python
# Делаем вывод, что плоскость может быть построена, но с другим ядром
clf = SVC(kernel = 'rbf', C = 10)
clf.fit(X, y)
fig = plt.figure(figsize = (10, 10), dpi = 150)
plt.scatter(X[:, 0], X[:, 1], c = y, s = 50, cmap = 'winter')
plot_svc_decision_function(clf)
```
